```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
botData = pd.read_csv('bot_bids.csv')
humData = pd.read_csv('hum_bids.csv')
trainData = pd.read_csv('train.csv')
```

```python
#len(botData.index)
feature = 'url' #auction, merchandise, device, country, url
botUrl = botData.groupby(feature).count().sort('bid_id', ascending=False) / len(botData.index)
botUrl.head(10)
```

```python
humUrl = humData.groupby(feature).count().sort('bid_id', ascending=False) / len(humData.index)
humUrl.head(10)
```

```python
from sets import Set

#analyze payment account
#payment = trainData.get(['payment_account', 'address', 'outcome']).values
paymentArr = np.array([])
addressArr = np.array([])
outcomeArr = np.array([])

paymentSet = Set({})
cutSize = 6

for elem in trainData.values:
    currPayment = elem[1]
    currPaymentCut = currPayment[0:len(currPayment)-cutSize]
    currAddr = elem[2]
    currAddrCut = currAddr[0:len(currAddr)-cutSize]
    classification = elem[3]
    paymentArr = np.append(paymentArr, currPaymentCut)
    addressArr = np.append(addressArr, currAddrCut)
    outcomeArr = np.append(outcomeArr, classification)

#convert to pandas dataframe        
filteredDF = pd.DataFrame({'payment_account':paymentArr, 'address':addressArr, 'outcome': outcomeArr})
filteredDF.head(5)
```

```python
payCntDF = filteredDF.groupby('payment_account').count()
paySumDF = filteredDF.groupby('payment_account').sum()
```
